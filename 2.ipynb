```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')

print("Dataset Info:")
display(df.info())
print("\nMissing Values:")
display(df.isnull().sum())

display(df.head())
```

```python
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])

df.drop(columns=['deck', 'embark_town', 'alive', 'class', 'who'], inplace=True)

print("Cleaned Data Summary:")
display(df.describe())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

sns.barplot(x='sex', y='survived', data=df, ax=axes[0], palette='muted', hue='sex', legend=False)
axes[0].set_title('Survival Rate by Gender')

sns.barplot(x='pclass', y='survived', hue='sex', data=df, ax=axes[1], palette='magma')
axes[1].set_title('Survival Rate by Class and Gender')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='survived', multiple='stack', kde=True, palette='coolwarm')
plt.title('Age Distribution by Survival Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
